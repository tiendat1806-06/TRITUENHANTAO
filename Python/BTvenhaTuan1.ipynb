```python
def tron_hai_mang(mang_a, mang_b):
    mang_ket_qua = []

    do_dai_min = min(len(mang_a), len(mang_b))

    for i in range(do_dai_min):
        gia_tri_min = min(mang_a[i], mang_b[i])
        mang_ket_qua.append(gia_tri_min)

    if len(mang_a) > do_dai_min:
        mang_ket_qua.extend(mang_a[do_dai_min:])
    elif len(mang_b) > do_dai_min:
        mang_ket_qua.extend(mang_b[do_dai_min:])

    return mang_ket_qua

# Nhập dữ liệu từ bàn phím
mang_a = list(map(int, input("Nhập mảng a (các số cách nhau bằng dấu cách): ").split()))
mang_b = list(map(int, input("Nhập mảng b (các số cách nhau bằng dấu cách): ").split()))

ket_qua = tron_hai_mang(mang_a, mang_b)

print(f"Mảng a: {mang_a}")
print(f"Mảng b: {mang_b}")
print(f"Mảng kết quả: {ket_qua}")
```

```python
import math

def nhap_ma_tran():
    m = int(input("Nhập số dòng: "))
    n = int(input("Nhập số cột: "))
    a = []
    for i in range(m):
        row = []
        for j in range(n):
            val = int(input(f"Nhập phần tử a[{i}][{j}]: "))
            row.append(val)
        a.append(row)
    return a, m, n

def cau_a_tong_tam_giac_tren_cheo_phu(a, n):
    tong = 0
    for i in range(n):
        for j in range(n):
            if i + j <= n - 1:
                tong += a[i][j]
    return tong

def cau_b_tri_tuyet_doi(a, m, n):
    for i in range(m):
        for j in range(n):
            a[i][j] = abs(a[i][j])
    return a

def cau_c_thay_chan_bang_x(a, m, n, x):
    for i in range(m):
        for j in range(n):
            if a[i][j] % 2 == 0:
                a[i][j] = x
    return a

def cau_d_kiem_tra_toan_chan(a, m, n):
    for i in range(m):
        for j in range(n):
            if a[i][j] % 2 != 0:
                return False
    return True

def cau_e_kiem_tra_doi_xung(a, n):
    for i in range(n):
        for j in range(i + 1, n):
            if a[i][j] != a[j][i]:
                return False
    return True

def cau_f_duong_cheo_chinh_tang_dan(a, n):
    for i in range(n - 1):
        if a[i][i] >= a[i+1][i+1]:
            return False
    return True

def cau_g_xuat_tam_giac_duoi_cheo_phu(a, n):
    print("Các phần tử thuộc tam giác dưới đường chéo phụ:")
    for i in range(n):
        for j in range(n):
            if i + j >= n - 1:
                print(a[i][j], end="\t")
        print()

def cau_h_duong_cheo_phu_giam_dan(a, n):
    for i in range(n - 1):
        if a[i][n - 1 - i] <= a[i+1][n - 1 - (i+1)]:
            return False
    return True

def in_ma_tran(a):
    for row in a:
        print("\t".join(map(str, row)))

def menu():
    m, n = 3, 3
    a = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

    while True:
        print("\n" + "="*30)
        print("      MENU BÀI TẬP MA TRẬN")
        print("="*30)
        print("Ma trận hiện tại:")
        in_ma_tran(a)
        print("-" * 30)
        print("1. Nhập ma trận mới")
        print("a. Tính tổng tam giác trên đường chéo phụ")
        print("b. Chuyển phần tử âm thành trị tuyệt đối")
        print("c. Thay phần tử chẵn bằng số x")
        print("d. Kiểm tra ma trận toàn chẵn")
        print("e. Kiểm tra ma trận đối xứng")
        print("f. Kiểm tra đường chéo chính tăng dần")
        print("g. Xuất tam giác dưới đường chéo phụ")
        print("h. Kiểm tra đường chéo phụ giảm dần")
        print("0. Thoát")

        chon = input("Mời bạn chọn chức năng: ").lower()

        if chon == '0':
            break
        elif chon == '1':
            a, m, n = nhap_ma_tran()
        elif chon == 'a':
            if m == n: print(f"Tổng: {cau_a_tong_tam_giac_tren_cheo_phu(a, n)}")
            else: print("Yêu cầu ma trận vuông!")
        elif chon == 'b':
            a = cau_b_tri_tuyet_doi(a, m, n)
        elif chon == 'c':
            x = int(input("Nhập x: "))
            a = cau_c_thay_chan_bang_x(a, m, n, x)
        elif chon == 'd':
            print("Toàn chẵn" if cau_d_kiem_tra_toan_chan(a, m, n) else "Có số lẻ")
        elif chon == 'e':
            if m == n: print("Đối xứng" if cau_e_kiem_tra_doi_xung(a, n) else "Không đối xứng")
            else: print("Yêu cầu ma trận vuông!")
        elif chon == 'f':
            if m == n: print("Tăng dần" if cau_f_duong_cheo_chinh_tang_dan(a, n) else "Không tăng")
            else: print("Yêu cầu ma trận vuông!")
        elif chon == 'g':
            if m == n: cau_g_xuat_tam_giac_duoi_cheo_phu(a, n)
            else: print("Yêu cầu ma trận vuông!")
        elif chon == 'h':
            if m == n: print("Giảm dần" if cau_h_duong_cheo_phu_giam_dan(a, n) else "Không giảm")
            else: print("Yêu cầu ma trận vuông!")

if __name__ == "__main__":
    menu()
```
