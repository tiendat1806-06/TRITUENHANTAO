```python
def tron_mang(a, b):
    ket_qua = []
    i, j = 0, 0

    while i < len(a) and j < len(b):
        ket_qua.append(a[i] + b[j])
        i += 1
        j += 1

    ket_qua.extend(a[i:])
    ket_qua.extend(b[j:])

    return ket_qua

# Nhập mảng a
n = int(input("Nhập số phần tử mảng a: "))
a = [int(input(f"a[{i}] = ")) for i in range(n)]

# Nhập mảng b
m = int(input("Nhập số phần tử mảng b: "))
b = [int(input(f"b[{i}] = ")) for i in range(m)]

print("Mảng a:", a)
print("Mảng b:", b)
print("Mảng kết quả:", tron_mang(a, b))
```

```python
import random

def tao_mang(m, n):
    return [[random.randint(1, 100) for j in range(n)] for i in range(m)]

def in_mang(a):
    print("\nMảng a:")
    for dong in a:
        print(dong)

def xuat_dong(a, k):
    print(f"Dòng {k}:", a[k])

def xuat_cot(a, k):
    print(f"Cột {k}:", [a[i][k] for i in range(len(a))])

def dong_tong_lon_nhat(a):
    tong_max = sum(a[0])
    dong_max = 0
    for i in range(len(a)):
        if sum(a[i]) > tong_max:
            tong_max = sum(a[i])
            dong_max = i
    print(f"Dòng {dong_max} có tổng lớn nhất: {tong_max}")

def cot_tich_nho_nhat(a):
    cot_min, min_tich = 0, None
    for j in range(len(a[0])):
        tich = 1
        for i in range(len(a)):
            tich *= a[i][j]
        if min_tich is None or tich < min_tich:
            min_tich = tich
            cot_min = j
    print(f"Cột {cot_min} có tích nhỏ nhất: {min_tich}")

def xuat_dong_chan_cot_le(a):
    print("Phần tử dòng chẵn, cột lẻ:")
    for i in range(len(a)):
        if i % 2 == 0:
            for j in range(len(a[i])):
                if j % 2 != 0:
                    print(f"  a[{i}][{j}] = {a[i][j]}")

def tb_chan_dong_le(a):
    tong, dem = 0, 0
    for i in range(len(a)):
        if i % 2 != 0:
            for x in a[i]:
                if x % 2 == 0:
                    tong += x
                    dem += 1
    print("TB chẵn ở dòng lẻ:", tong / dem if dem > 0 else 0)

def tb_bien(a):
    tong, dem = 0, 0
    for i in range(len(a)):
        for j in range(len(a[i])):
            if i == 0 or i == len(a)-1 or j == 0 or j == len(a[i])-1:
                tong += a[i][j]
                dem += 1
    print("TB biên:", tong / dem if dem > 0 else 0)

def tb_tich_khong_bien(a):
    tich, dem = 1, 0
    for i in range(1, len(a)-1):
        for j in range(1, len(a[i])-1):
            tich *= a[i][j]
            dem += 1
    print("TB tích không biên:", tich / dem if dem > 0 else 0)

# ========== MENU ==========
def menu():
    m = int(input("Nhập số dòng m: "))
    n = int(input("Nhập số cột n: "))
    a = tao_mang(m, n)
    in_mang(a)

    while True:
        print("\n========== MENU ==========")
        print("a. Tạo mảng mới")
        print("b. Xuất dòng k")
        print("c. Xuất cột k")
        print("d. Tìm dòng có tổng lớn nhất")
        print("e. Tìm cột có tích nhỏ nhất")
        print("f. Xuất phần tử dòng chẵn, cột lẻ")
        print("g. TB cộng phần tử chẵn ở dòng lẻ")
        print("h. TB cộng phần tử thuộc biên")
        print("i. TB tích phần tử không thuộc biên")
        print("0. Thoát")
        print("==========================")

        chon = input("Chọn chức năng: ").strip().lower()

        if chon == "a":
            m = int(input("Nhập số dòng m: "))
            n = int(input("Nhập số cột n: "))
            a = tao_mang(m, n)
            in_mang(a)
        elif chon == "b":
            k = int(input("Nhập k: "))
            xuat_dong(a, k)
        elif chon == "c":
            k = int(input("Nhập k: "))
            xuat_cot(a, k)
        elif chon == "d":
            dong_tong_lon_nhat(a)
        elif chon == "e":
            cot_tich_nho_nhat(a)
        elif chon == "f":
            xuat_dong_chan_cot_le(a)
        elif chon == "g":
            tb_chan_dong_le(a)
        elif chon == "h":
            tb_bien(a)
        elif chon == "i":
            tb_tich_khong_bien(a)
        elif chon == "0":
            print("Thoát chương trình!")
            break
        else:
            print("Chọn không hợp lệ, thử lại!")

menu()
```

```python
import random
import datetime

def tao_sinh_vien(msv, ten, nam_sinh, dtb):
    return {"msv": msv, "ten": ten, "nam_sinh": nam_sinh, "dtb": dtb}

def in_sinh_vien(sv):
    print(f"MSV: {sv['msv']} | Tên: {sv['ten']} | Năm sinh: {sv['nam_sinh']} | ĐTB: {sv['dtb']}")

def tao_mang_sv(n):
    mang = []
    for i in range(n):
        print(f"\nNhập sinh viên thứ {i+1}:")
        msv = input("  Mã SV (10 ký tự): ")[:10]
        ten = input("  Tên (tối đa 20 ký tự): ")[:20]
        nam_sinh = int(input("  Năm sinh: "))
        dtb = float(input("  Điểm trung bình: "))
        mang.append(tao_sinh_vien(msv, ten, nam_sinh, dtb))
    return mang

def in_mang_sv(mang):
    print("\nDanh sách sinh viên:")
    for sv in mang:
        in_sinh_vien(sv)

def dem_len_lop(mang):
    dem = sum(1 for sv in mang if sv['dtb'] >= 5)
    print(f"Số sinh viên đủ điều kiện lên lớp: {dem}")

def xuat_du_20_tuoi(mang):
    nam_hien_tai = datetime.datetime.now().year
    print("Sinh viên đủ 20 tuổi:")
    for sv in mang:
        if nam_hien_tai - sv['nam_sinh'] >= 20:
            in_sinh_vien(sv)

def dem_he_dh(mang):
    dem = sum(1 for sv in mang if sv['msv'][2:4] == "DH")
    print(f"Số sinh viên hệ đại học: {dem}")

def dem_ten_lan(mang):
    dem = sum(1 for sv in mang if sv['ten'].split()[-1] == "Lan")
    print(f"Số sinh viên có tên Lan: {dem}")

def dem_ho_phan(mang):
    dem = sum(1 for sv in mang if sv['ten'].split()[0] == "Phan")
    print(f"Số sinh viên có họ Phan: {dem}")

def menu():
    mang = []

    while True:
        print("\n========== MENU SINH VIÊN ==========")
        print("a. Tạo mảng sinh viên")
        print("b. In danh sách sinh viên")
        print("c. Đếm sinh viên đủ điều kiện lên lớp")
        print("d. Xuất sinh viên đủ 20 tuổi")
        print("e. Đếm sinh viên hệ đại học")
        print("f. Đếm sinh viên có tên Lan")
        print("g. Đếm sinh viên có họ Phan")
        print("0. Thoát")
        print("=====================================")

        chon = input("Chọn chức năng: ").strip().lower()

        if chon == "a":
            n = int(input("Nhập số sinh viên: "))
            mang = tao_mang_sv(n)
            in_mang_sv(mang)
        elif chon == "b":
            if mang:
                in_mang_sv(mang)
            else:
                print("Chưa có dữ liệu, chọn a để nhập!")
        elif chon == "c":
            dem_len_lop(mang)
        elif chon == "d":
            xuat_du_20_tuoi(mang)
        elif chon == "e":
            dem_he_dh(mang)
        elif chon == "f":
            dem_ten_lan(mang)
        elif chon == "g":
            dem_ho_phan(mang)
        elif chon == "0":
            print("Thoát chương trình!")
            break
        else:
            print("Chọn không hợp lệ, thử lại!")

menu()
```
